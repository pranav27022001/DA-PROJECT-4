```python
figure = exp.bar(df, 
                x=df["venue"],
                 title="venues where most matches were held")
figure.show()
```

So we can find that SCG, MCG and Adelaide oval hosted most of the matches with 7 matches respectively

```python
figure = exp.bar(df, 
                x=df["top scorer"], 
                y = df["highest score"], 
                color = df["highest score"],
                title="Top Scorers in t20 World Cup 2022")
figure.update_xaxes(tickangle=90)
figure.show()
```

From the above analysis we can figure out that (Rilee Rossow) is the highest run scorer in the complete tournament


```python
df.head(2)
```

```python
fig = exp.bar(df, x=df['winner'], title="Number of Matches Won by teams in t20 World Cup 2022")
fig.show()
```

So from the above analysis we can see that england won the most no of matches with 5 and india and pakistan with 4 respectively

```python
df.columns
```

```python
matches_won_by = df['won by'].value_counts()
matches_won_by = pd.DataFrame(matches_won_by)
matches_won_by.reset_index(inplace=True)
matches_won_by
```

```python
toss_decision = df['toss decision'].value_counts()
```

```python
label = toss_decision.index
value = toss_decision.values
total = sum(value)
plt.pie(value, labels = label, startangle=90, autopct=lambda p: '{:.0f}%'.format(p * total / 100), shadow=True, colors=['red', 'yellow'])
plt.show()
```

From this we can analyse that after winning toss most of the teams decided to bat first

```python
figure = exp.bar(df, 
                x = df["player of the match"], 
                title="Player of the Match Awards in t20 World Cup 2022")
figure.update_xaxes(tickangle=90)
figure.show()
```

From the above analysis we can analyse that only ***Sam Curran, Virat Kohli, Taskin Ahmed, Suryakumar Yadav and Shadab khan*** got 2 man of the match award, rest of them are only with one Man of the match award

```python
figure = exp.bar(df, 
                x=df["best bowler"],
                title="Best Bowlers in t20 World Cup 2022")
figure.update_xaxes(tickangle=90)
figure.show()
```

So we all know that as England won the world cup. Reason behind it was the overall perfeormance with bat and the ball. As we can see with ball ***Sam Curran*** was the best bowler in entire tournament.
